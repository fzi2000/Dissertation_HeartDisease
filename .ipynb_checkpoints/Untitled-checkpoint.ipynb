```python
# üöÄ Import Required Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from imblearn.over_sampling import SMOTE
from sklearn.metrics import accuracy_score, recall_score, f1_score, roc_auc_score, classification_report, confusion_matrix, ConfusionMatrixDisplay

# üöÄ Load Dataset
df = pd.read_csv("data/framingham.csv")  # Update the path if needed

# üöÄ Data Preprocessing
# Fill missing values with the mean
imputer = SimpleImputer(strategy="mean")
df_imputed = pd.DataFrame(imputer.fit_transform(df), columns=df.columns)

# Standardize numerical features
scaler = StandardScaler()
X = df_imputed.drop(columns=["TenYearCHD"])
X_scaled = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)
y = df_imputed["TenYearCHD"]

# üöÄ Split the Dataset
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42, stratify=y)

# üöÄ Apply SMOTE for Class Balancing
smote = SMOTE(sampling_strategy=1.0, random_state=42)
X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)

# üöÄ Define ANN Model
def create_ann_model(input_shape):
    model = Sequential([
        Dense(256, activation='relu', input_shape=(input_shape,)),
        BatchNormalization(),
        Dropout(0.3),
        
        Dense(128, activation='relu'),
        BatchNormalization(),
        Dropout(0.3),

        Dense(64, activation='relu'),
        BatchNormalization(),
        Dropout(0.2),

        Dense(32, activation='tanh'),
        Dense(1, activation='sigmoid')  # Output Layer for Binary Classification
    ])
    
    # Compile Model
    model.compile(optimizer=Adam(learning_rate=0.0003),  
                  loss='binary_crossentropy', 
                  metrics=['accuracy'])
    
    return model

# üöÄ Create ANN Model
input_shape = X_train_smote.shape[1]  
ann_model = create_ann_model(input_shape)

# üöÄ Early Stopping (to prevent overfitting)
early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

# üöÄ Train the ANN Model
history = ann_model.fit(X_train_smote, y_train_smote, epochs=100, batch_size=64, validation_data=(X_test, y_test),
                         verbose=1, callbacks=[early_stopping])

# üöÄ Make Predictions
y_pred_prob = ann_model.predict(X_test)
y_pred = (y_pred_prob >= 0.5).astype(int)  # Convert Probabilities to Binary Labels

# üöÄ Compute Performance Metrics
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
auc = roc_auc_score(y_test, y_pred_prob)

# üöÄ Print Performance Report
print("\nüìä **Optimized ANN Performance:**")
print(classification_report(y_test, y_pred))

print(f"‚úÖ **Accuracy:** {accuracy * 100:.2f}%")
print(f"‚úÖ **Precision:** {precision:.2f}")
print(f"‚úÖ **Recall:** {recall:.2f}")
print(f"‚úÖ **F1 Score:** {f1:.2f}")
print(f"‚úÖ **AUC-ROC:** {auc:.2f}")

# üöÄ Plot Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["No CHD", "CHD"]).plot(cmap="Blues")
plt.title("Confusion Matrix - ANN Model")
plt.show()
```

```python
# üöÄ Import Required Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization, LeakyReLU
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from imblearn.over_sampling import SMOTE
from sklearn.metrics import accuracy_score, recall_score, f1_score, roc_auc_score, classification_report, confusion_matrix, ConfusionMatrixDisplay

# üöÄ Load Dataset
df = pd.read_csv("data/framingham.csv")  # Ensure you have the correct path

# üöÄ Data Preprocessing
# Fill missing values using Mean Imputation
imputer = SimpleImputer(strategy="mean")
df_imputed = pd.DataFrame(imputer.fit_transform(df), columns=df.columns)

# Standardize Numerical Features
scaler = StandardScaler()
X = df_imputed.drop(columns=["TenYearCHD"])
X_scaled = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)
y = df_imputed["TenYearCHD"]

# üöÄ Split Dataset
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42, stratify=y)

# üöÄ Apply SMOTE for Class Balancing
smote = SMOTE(sampling_strategy=1.0, random_state=42)
X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)

# üöÄ Define Deep Learning Model (Fully Connected Neural Network - FCNN)
def create_fcnn_model(input_shape):
    model = Sequential([
        Dense(512, input_shape=(input_shape,)),
        BatchNormalization(),
        LeakyReLU(alpha=0.1),
        Dropout(0.4),
        
        Dense(256),
        BatchNormalization(),
        LeakyReLU(alpha=0.1),
        Dropout(0.3),
        
        Dense(128),
        BatchNormalization(),
        LeakyReLU(alpha=0.1),
        Dropout(0.3),

        Dense(64),
        BatchNormalization(),
        LeakyReLU(alpha=0.1),
        Dropout(0.2),

        Dense(32),
        BatchNormalization(),
        LeakyReLU(alpha=0.1),
        Dropout(0.2),

        Dense(1, activation='sigmoid')  # Output Layer for Binary Classification
    ])
    
    # Compile Model with Adam Optimizer and Adaptive Learning Rate
    model.compile(optimizer=Adam(learning_rate=0.0002),  
                  loss='binary_crossentropy', 
                  metrics=['accuracy'])
    
    return model

# üöÄ Create and Train the Model
input_shape = X_train_smote.shape[1]  
fcnn_model = create_fcnn_model(input_shape)

# üöÄ Early Stopping to prevent overfitting
early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

# üöÄ Train the Model
history = fcnn_model.fit(X_train_smote, y_train_smote, epochs=120, batch_size=64, validation_data=(X_test, y_test),
                         verbose=1, callbacks=[early_stopping])

# üöÄ Make Predictions
y_pred_prob = fcnn_model.predict(X_test)
y_pred = (y_pred_prob >= 0.5).astype(int)  # Convert probabilities to binary labels

# üöÄ Compute Performance Metrics
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
auc = roc_auc_score(y_test, y_pred_prob)

# üöÄ Print Performance Report
print("\nüìä **Optimized FCNN Model Performance:**")
print(classification_report(y_test, y_pred))

print(f"‚úÖ **Accuracy:** {accuracy * 100:.2f}%")
print(f"‚úÖ **Precision:** {precision:.2f}")
print(f"‚úÖ **Recall:** {recall:.2f}")
print(f"‚úÖ **F1 Score:** {f1:.2f}")
print(f"‚úÖ **AUC-ROC:** {auc:.2f}")

# üöÄ Plot Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["No CHD", "CHD"]).plot(cmap="Blues")
plt.title("Confusion Matrix - FCNN Model")
plt.show()
```
